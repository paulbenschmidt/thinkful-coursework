```python
fuzzies = []
for title in df.journal_title.unique():
    for second_title in df.journal_title.unique():
        ratio = fuzz.ratio(title, second_title)
        if ratio > 90 and ratio < 100:
            fuzzies.append((title, second_title, ratio))
```

```python
matches = pd.DataFrame(fuzzies, columns=['title', 'second_title', 'ratio'])
```

```python
matches[matches.ratio > 92]
```

```python
df.journal_title = df.journal_title.apply(lambda title: (title.replace('the', '')
                                                        .replace('jnl', 'journal')
                                                        .replace('plosone', 'plos one')
                                                        .replace('1', 'one')
                                                        .replace('&', '')
                                                        .replace('and', '')
                                                        .replace('.', '')
                                                        .replace('developmental', 'development')
                                                        .replace('acids', 'acid')
                                                        .replace('jounral', 'journal')
                                                        .replace('biologicial', 'biological')
                                                        .replace('archiv', 'archive')
                                                        .replace('epidemiol', 'epidemiology')
                                                        .replace('angewande', 'angewandte')
                                                        ))
```

```python
df.cost = df.cost.str.strip().str.extract('(\d+\.\d\d)').astype('float')
```
