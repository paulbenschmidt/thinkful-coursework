```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Normal (Gaussian) Distribution
# normal([loc, scale, size])

mean = 0
sd = 5
n = 100

normal = np.random.normal(mean, sd, n)
plt.hist(normal)
plt.title('Normal (Gaussian) Distribution')
plt.axvline(normal.mean(), color='r', linestyle='solid')
plt.axvline(normal.mean()-normal.std(), color='r', linestyle='dashed')
plt.axvline(normal.mean()+normal.std(), color='r', linestyle='dashed')
plt.show()

print('''Evaluation: Useful. ''')
```

```python
# 2. Gamma Distribution
# gamma(shape[, scale, size])

shape = 1
scale = 1
size = 100

gamma = np.random.gamma(shape, scale, size)
plt.hist(gamma)
plt.title('Gamma Distribution')
plt.axvline(gamma.mean(), color='r', linestyle='solid')
plt.axvline(gamma.mean()-gamma.std(), color='r', linestyle='dashed')
plt.axvline(gamma.mean()+gamma.std(), color='r', linestyle='dashed')
plt.show()

print('''Evaluation: Not Useful.''')
```

```python
# 3. Binomial Distribution
# binomial(n, p[, size])

number_of_trials = 100
percentage_of_success = .75
size = 100

binomial = np.random.binomial(number_of_trials, percentage_of_success, size)
plt.hist(binomial)
plt.title('Binomial Distribution')
plt.axvline(binomial.mean(), color='r', linestyle='solid')
plt.axvline(binomial.mean()-binomial.std(), color='r', linestyle='dashed')
plt.axvline(binomial.mean()+binomial.std(), color='r', linestyle='dashed')
plt.show()

print('Evaluation: Useful.')
```

```python
# 4. Rayleigh Distribution
# rayleigh([scale, size])

scale = 1
size = 100

rayleigh = np.random.rayleigh(scale, size)
plt.hist(rayleigh)
plt.title('Rayleigh Distribution')
plt.axvline(rayleigh.mean(), color='r', linestyle='solid')
plt.axvline(rayleigh.mean()-rayleigh.std(), color='r', linestyle='dashed')
plt.axvline(rayleigh.mean()+rayleigh.std(), color='r', linestyle='dashed')
plt.show()

print("Evaluation: Not very useful (although it doesn't account for extending variance above mean.)")
```

```python
# Exponential Distribution
# exponential([scale, size])

scale = 1
size = 100

exponential = np.random.exponential(scale, size)
plt.hist(exponential)
plt.title("Exponential Distribution")
plt.axvline(exponential.mean(), color='r', linestyle='solid')
plt.axvline(exponential.mean()-exponential.std(), color='r', linestyle='dashed')
plt.axvline(exponential.mean()+exponential.std(), color='r', linestyle='dashed')
plt.show()

print('''Evaluation: Not Useful (while mean shows greatest density throughout the data, it doesn't show that the density is more accute when located less than mean.)''')
```

```python
# 6. Triangular Distribution
# triangular(left, mode, right[, size])

lower_limit = 0
mode = 75
upper_limit = 100
size = 100

triangular = np.random.triangular(lower_limit, mode, upper_limit, size)
plt.hist(triangular)
plt.title('Triangular')
plt.axvline(triangular.mean(), color='r', linestyle='solid')
plt.axvline(triangular.mean()-triangular.std(), color='r', linestyle='dashed')
plt.axvline(triangular.mean()+triangular.std(), color='r', linestyle='dashed')
plt.show()

print('Evaluation: Not Useful (the standard variation does not clearly show the peak as 75)')
```

```python
# Additionally:

normal1 = np.random.normal(5,.5,100)
normal2 = np.random.normal(10,1,100)
normal3 = normal1 + normal2

plt.hist(normal3)
plt.axvline(normal3.mean(), color='r', linestyle='solid')
plt.axvline(normal3.mean()-normal3.std(), color='r', linestyle='dashed')
plt.axvline(normal3.mean()+normal3.std(), color='r', linestyle='dashed')
plt.show()
```
