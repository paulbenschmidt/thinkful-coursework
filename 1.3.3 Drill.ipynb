## Drill 1: Choose one variable; plot four ways

```python
age_groups = df.groupby(['age']).agg('mean').drop(['year','HDI for year', 'gdp_per_capita ($)'], axis=1)
age_groups
```

```python
g = sns.catplot(x='age', y='suicides_no', data=df, kind='bar', height=6, aspect=2)
g.set_ylabels('Country-Suicide Count')
plt.show()

#Pros: clearly shows differences between variables
#Cons: does not show the distributions within each age group (only shows maximum count)
```

```python
sns.catplot(x='age', y='suicides_no', data=df, height=6, aspect=1.5, alpha=.5)

#Pros: shows actual distribution within each age group; shows distinctions between age groups
#Cons: doesn't convey the concentration at the bottom end of each bar
```

```python
sns.set(style="darkgrid")
bar = sns.catplot(x="sex", y="suicides_no", hue="age", data=df, height=6, aspect=2, kind="bar", palette="pastel", ci=95)
bar.despine(left=True)
bar.set_ylabels("Suicides")
bar.set_xlabels("")
plt.title('Barplot: Suicides by Sex and Age Group')
plt.show()

#Pros: differentiates between sexes and communicates larger male suicide rates
#Cons: youngest age group is negligible
```

```python
sns.set(style="whitegrid")
g = sns.catplot(x="sex", y="suicides_no", hue="age", data=df, height=6, aspect=2, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Suicides")
g.set_xlabels("")
plt.title('Pointplot: Meal Tips by Time and Gender')
plt.show()

#Pros: shows the consistency rates between age groups (instead of merely communicating that they are high or low)
#Cons: the visualized data of males diminishes value of visualizing female
```

```python
sns.set(style="ticks") 
g = sns.FacetGrid(df, row="age", col="sex")
g.map(plt.hist, "suicides_no", color="steelblue",  lw=1)
plt.subplots_adjust(top=.95)
g.set_axis_labels('')
plt.suptitle('Suicides by Gender and Age')
plt.show()

#To be honest, I don't know what the x-axis is communicating.
```

## Drill 2: Choose two continuous variables; plot three ways

```python
g = sns.lmplot(y='suicides_no',
               x='gdp_per_capita ($)',
               data=df,
               hue='sex',
               height=8,
               fit_reg=True,
              scatter_kws={'alpha':0.4})
g.set_ylabels("Suicides")
g.set_xlabels("GDP Per Capita")
plt.title('Scatterplot: Suicides by GDP Per Capita')
plt.show()

#Pros: shows variance between sexes regarding suicide is only primarily below 80000 GDP per capita,shows mysterious 60000 drop off
#cons: doesn't communicate density well
```

```python
sns.jointplot(x='gdp_per_capita ($)', y='suicides_no', data=df,
                   kind='reg', space=0, height=8, ratio=4)

#Pros: shows density
#Cons: doesn't show difference between sexes
```

```python
sns.set(style="white")
g = sns.lmplot(y='suicides_no', 
               x='gdp_per_capita ($)',
               hue='sex',
               data=df,
               col='age',
               col_wrap=3,
               fit_reg=False,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Suicides")
g.set_xlabels("GDP Per Capita")
plt.suptitle('Scatterplot Regression: Suicides by GDP Per Capita, split by Sex')
plt.subplots_adjust(top=0.9)
plt.ylim(0, None)
plt.xlim(-5000, 150000)
plt.show()

#Pros: shows higher deaths in different age groups
#Cons: doesn't communicate density
```

## Drill 3: Choose one continuous variable and one categorical variable; plot six ways

```python
sns.catplot(x="age", y="suicides_no", hue="sex",
            kind="violin", split=True, data=df, height=7, aspect=2)

#Pros: Shows that high suicide rates are outlier countries
#Cons: Doesn't really give an accurate base since much of the data goes below 0
```

```python
sns.catplot(x='age', y='suicides_no', hue='sex', data=df, kind='bar', height=6, aspect=2)

#Pros: shows difference in counts clearly between the categorical variable
#Cons: gives count across countries instead of showing distrubtion
```

```python
sns.relplot(x='year', y='suicides_no', hue='sex', style='sex', kind='line', data=df, height=8)

#Pros: Clearly shows that suicide rate across years is consistently higher in one categorical variable
#Cons: Doesn't factor individual countries but instead groups them all by year
```

```python
sns.relplot(x="year", y="suicides_no", hue="sex", data=df, height=8)

#Pros: shows individual countries, shows that above 5000 is consistently male
#Cons: hard to interpret, data below 5000 doesn't communicate important information
```

```python
sns.relplot(x="population", y="suicides_no", hue="sex", data=df, height=8)

#Pros: shows interesting correlation between male/female suicides for larger populations, shows interesting spike in 2e7 population
#Cons: left-bottom quadrant is too loud, doesn't take into account percentage of suicides to total population
```

```python
sns.catplot(x='age', y='suicides_no', hue='sex', data=df, kind='boxen', height=6, aspect=2)

#Pros: strongly communicates the outliers and density across age groups
#Cons: because 35-54 age group is so large, the other data is difficult to read
```
